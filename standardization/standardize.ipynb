```python
def extract_content(content):

    title_pattern = r"/'title':\s*'([^']*)'/gm"
    question_pattern = r"'question':\s*'([^']*)'"
    answer_pattern = r"'answer':\s*'([^']*)'"

    title_match = re.search(title_pattern, content)
    question_match = re.search(question_pattern, content)
    answer_match = re.search(answer_pattern, content)

    title_content = title_match.group(1) if title_match else None
    question_content = question_match.group(1) if question_match else None
    answer_content = answer_match.group(1) if answer_match else None

    return title_content, question_content, answer_content


test_title, test_question, test_answer = extract_content((training_df[["javascript_content"]].iloc[0].to_string()))
print(test_question)
```
